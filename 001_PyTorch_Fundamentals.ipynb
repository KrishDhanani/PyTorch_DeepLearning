# Tensor datatypes

**Note**: Tensor datatypes is one of the 3 big errors you'll run into with PyTorch & deep learning:
1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on right Device

```
# Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=torch.float32,   # What datatype is the tensor (e.g. float32 or float64) # it healps in Precision computing
                               device=None,   # What device is your tensor on
                               requires_grad=False)   # Whether or not to track gradients with this tensors operations
float_32_tensor
```

```
float_32_tensor.dtype
```

```
float_16_tensor = float_32_tensor.type(torch.float16)     # We converted the float 32 tensor to float 16 tensor.
float_16_tensor
```

```
int_32_tensor = torch.tensor([3, 6, 9], dtype=torch.long)
int_32_tensor
```

# Matrix Multiplication

Two main ways of performing multiplication in neural networks and deep learning:

1. Element-wise multiplication
2. Matrix multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
* `(3,2) @ (3,2)` won't work
* `(2,3) @ (3,2)` will work
* `(3,2) @ (2,3)` will work

```
# 1. Elementwise Multiplication
tensor = torch.randint(low=1, high=5, size=[3])

print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```
# 2. Matrix multiplication [dot product]
torch.matmul(tensor, tensor)
```

```
tensor @ tensor
```

```
torch.matmul(input=torch.rand(3,2), other=torch.rand(2,3))
```

```
%%time
torch.matmul(tensor,tensor)
```

# One of the most common errors in deep learning: shape errors

```
# Shape for matrix multiplication

tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])

# torch.mm(tensor_A, tensor_B)   # torch.mm is same as torch.matmul (it's an alice for matmul)
torch.matmul(tensor_A, tensor_B)
```

```
tensor_A.shape , tensor_B.shape
```

# Transpose

To fix out tensor shape issue, we can manipulate the shape of our tensors using a **transponse**.

A **transponse** switches the axes or dimensions of a given tensor.

```
tensor_B, tensor_B.shape
```

```
tensor_B.T, tensor_B.T.shape
```

```
# The matrix multiplication works when tensor_B is transposed

print(f"Original shape: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shape: tensor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print(f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape} <- inner dimensions must match")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

# Reshaping, Staking, Squeezing and unsqueezing tensors

- Reshaping - reshapes an input tensor to a defined shape
- View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
- Staking - combine multiple tensors on top of each othe (vstack) or side by side (hstack)
- Squeeze -remove all `1` dimensions from a tensor
- Unsqueeze - add a `1` dimension to a target tensor
- Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```
# Lets create tensor
import torch

x = torch.arange(1,11)
x, x.shape
```

```
# Add an extra dimension
x_reshaped = x.reshape(5,2)
x_reshaped, x_reshaped.shape
```

```
# Change the View : view is just sharing original memory of tensor
z = x.view(1, 10)
z, z.shape
```

```
# Changing z changes x (happening of sharing same memory)
z[:, 0] = 5
z, x
```

```
# Stack tensoprs on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```
# Squeeze and Unsqueeze

# torch.Squeeze(): removes all single dimensions from a target tensor
y = torch.arange(1,9)
y, y.shape  # Previous tensor looks and shape
```

```
y_squeezed = y.squeeze()   # After Squeeze looks like
y_squeezed
```

```
y.squeeze().shape   # After Squeezing shape
```

```
# torch.unsqueeze() - adds a single dimensions to a target at a specific dim (dimensions)
```

```
y_unsqueezed = y_squeezed.unsqueeze(dim=0)
y_unsqueezed
```

```
# Permute: torch.permute() -> It's kind of rearanging the tensor dimension
# Also it's kind of view of original tensor like sharing the same memory like view methode
# Using this thing for images
x = torch.randn(2, 3, 5)
x.shape
```

```
permuted_x = torch.permute(x, (2, 0, 1)).shape   # It's taking firstly tensor then by the indaxing of shape of reshaping the original shape
permuted_x
```

```
print(f"Initial value from original tensor x at index 0,0,0: {x[0, 0, 0]}")
x[0,0,0] = 1
permuted_x = torch.permute(x, (2, 0, 1)) # Re-assign permuted_x to be the actual permuted tensor
print(f"After changing value from original tensor x at index 0,0,0: {permuted_x[0, 0, 0]}")
```
